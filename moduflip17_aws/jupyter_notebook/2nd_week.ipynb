### 1. 스위치의 개념과 스위치의 종류(L2,L3,L4,L7)에 대하여정리


#### 1-1. 스위치란?

스위치는 여러 컴퓨터를 연결하는 네트워크 분배기 역할을 한다. 컴퓨터나 모니터등의 전원 케이블을 연결할 때 전기 콘센트가 부족할 경우 '멀티탭'이라는 전기 용품을 사용하는 것처럼 네트워크 장비 중에서 허브(hub)와 스위치(switch)도 네트워크를 분배해주는 역할을 한다. 하나의 네트워크 라인에 여러대의 컴퓨터의 랜케이블을 꼽을 수 있도록 해주는 분배장비이다.

다만 허브와 스퀴치가 다른 점은 연결된 기기 중 하나에서 전송된 패킷이 허브에 연결된 모든 기기로 브로드 캐스팅된다는 점이다. 반면에 스위치는 패킷의 목적지 주소로 지정된 기기로 이어진 포트로만 패킷이 전달된다. 또한 스위치는 일반적으로 LAN 세그먼트와 세그먼트에 연결된 허브를 연결한다. 스위치는 동일한 LAN 세그먼트의 기기로 향하는 트래픽을 걸러내는 기능 덕분에 스위치는 자체 프로세싱 리소스와 네트워크 대역폭을 효율적으로 활용한다는 장점이 있다.

#### 1-2. OSI7 Layer
- 국제표준화기구(ISO)에서 개발한 모델로 컴퓨터 네트워크 프로토콜 디자인과 통신을 계층(Layer)로 나눠서 설명한 것이다. 일반적으로 OSI 7 layer라고 불리우고 다음과 같이 7단계로 나누어진다.  

__a. L1(Physical layer):__ 실제 장비들을 연결하기 위해 전기적 물리적 세부사항들을 정의한다.  
    ex) 핀들의 배치나 전압, 전선의 명세, hub1이나 리피터 L1이 이 계층에 포함된다.  

__b. L2(Data Link layer):__ Data Link layer는 point to point 간 신뢰성 있는 전송을 보장하기 위한 계층으로 CRC 기반의 오류제어와 흐름제어가 필요하다. 주소 값은 물리적으로 할당 받으며, 네트워크 카드 설정할때 사용되는 MAC Address가 여기에 해당한다  
    ex) 이더넷2, 네트워크 브릿지
    
__c. L3(Network layer):__ 여러 노드를 거칠때 경로를 찾아주는 역할을 해주는 계층으로 데이터를 네트워크를 통해서 전달하고, 그 과정에서 QoS를 제공하기 위한 기능적 절차적 수단을 제공한다. 라우팅, 흐름제어, 세그먼테이션, 오류제어, 인터네트워킹 등을 수행한다. 
    ex) 라우터, L3 스위치..

__d. L4(Transport Layer):__ End to End의 사용자들의 신뢰성 있는 데이터를 주고 받을 수 있도록 해준다. 시퀀스 넘버 기반의 오류제어 방식을 사용하며, 특정 연결의 유효성을 제어한다. L4 Layer의 일부 프로토콜은 상태 개념이 있고 (Stateful), 연결기반(Connection Oriented)이다. 이는 Transport Layer으 패킷들의 전송이 유효한지 확인하고 전송이 실패한 패킷들을 다시 전송한다는 것을 의미한다. 
    ex) TCP
    
__e. L5(Session Layer):__ 양 끝단의 응용 프로세스가 통신을 관리하기 위한 방법을 제공한다. 동시 송수신 방식(Duplex), 반이중방식(half Duplex), 전 이중방식(Full Duplex)등의 통신과 함께, 체크 포인팅과 유휴, 종료, 다시 시작과정 등을 수행한다. 이 계층은 TCP/IP 세션을 만들고 없애는 책임을 진다. 

__f. L6(Presentation Layer):__  코드간의 번역을 담당하여 사용자 시스템에서 데이터의 형식상 차이를 다루는 부담을 application layer로부터 덜어주는 역할을 한다. 
    ex) MIME 인코딩이나 암호화

__g. L7(Application Layer):__ 응용 프로세스와 직접 관계해서 서비스를 수행한다. ex) Telnet

#### 1-3. 스위치의 종류

|스위치 종류|역할|
|-|-|
|L2 스위치|OSI Layer2에 속하는 장비로 MAC Address를 참조하여 스위칭|
|L3 스위치|OSI Layer3에 속하는 장비로 IP주소를 참조하여 스위칭|
|L4 스위치|OSI Layer3~4에 속하는 장비로 IP주소 및 TCP/UDP 포트정보를 참조하여 스위칭|
|L7 스위치|OSI Layer3~7에 속하는 장비로 IP주소 및 TCP/UDP 포트정보 및 패킷 내용까지 참조하여 스위칭|

- L2 스위치
 스위치의 기본 정의에 가장 부합하느 것이 L2스위치이다. 따라서 L2스위치를 그냥 스위치라고 부르기도 한다. L2스위치는 허브에서 한단계 진화한 것으로 도착한 패킷의 제일 앞에 있는 목적지 MAC Address를 보고 어떤 포트로 보낼 것인지 판단한다.
 
- L3 스위치
 L3 스위치는 OSI7 Layer 중 3단계인 Network Layer에서 동작하므로 IP스위칭이 가능하다. 자신에게 온 패킷의 목적지가 외부에 존재하는 IP일 경우 그 패킷을 외부에 연결된 라우터로 보내줄 수 있다. 즉, L3 스위치는 라우터 기능도 갖고 있기 때문에 라우터와의 경계가 모호해진다.
 
- L4 스위치
 L4 Switch는 L4에서 동작한다. L4단계의 프로토콜인 TCP/UDP등에서 스위칭을 수행하므로 TCP와 UDP등의 헤더를 보고 그것이 FTP인지 HTTP인지 아니면 SMTP인지를 구분한 후 어떤 것을 우선시해서 스위칭 할지 판단할 수 있다. 로드밸런싱이 가능해진다는 것으로 이는 L3 스위치 이하에서는 불가능한 것이다.

- L7 스위치
 L7 스위치는 OSI7 Layer 중 상위의 Layer인 Application Layer에서 동작하는 스위치로 상당히 다양한 조건으로 로드벨런싱이 가능하다.
 ex) 이메일의 제목이나 문자열을 보고 내용을 파악한다거나 HTTP의 URL, FTP의 파일명, 쿠키정보, 특정 바이러스의 패턴등을 분석해서 보안에 유리하고 정교한 로드 밸런싱을 가능하게 해준다.
 


### 2. RAID의 개념과 RAID 구성방식에 대하여 정리(RAID0,1,4,5,6,1+0,0+1)

#### 2-1. RAID(Redundant Array of Independent Disks 혹은 Redundant Array of Inexpensive Disks)의 개념

- 여러개의 하드디스크에 중복된 데이터를 나눠서 저장하는 기술이다. 디스크 어레이라고도 한다. 데이터를 나누는 다양한 방법이 존재하며, 이 방법들을 레벨이라고 하는데 레벨에 따라서 저장장치의 신뢰성을 높이거나 전체적인 성능을 향상시키는 등 다양한 목적을 만족시킬 수 있다.  

- RAID는 여러개의 디스크를 하나로 묶어 하나의 논리적 디스크로 작동하게 하는데 하드웨어적인 방법과 소프트웨어적인 방법등이 있다. 하드웨어적인 방법은 운영체제에 이 디스크가 하나의 디스크처럼 보이게 한다. 소프트웨어적인 방법은 운영체제 안에서 구현되며, 사용자에게 디스크를 하나의 디스크처럼 보이게 한다.

#### 2-2. RAID 0 (striping)

- 배열드라이브에 데이터를 분산시킨다. 작업부하가 각 드라이브로 병렬화되고 분산되어서 여러 환경에서 일기 및 쓰기 처리량이 개선된다. 스트라이핑 모드를 쓰려며 최소한 2개의 드라이브가 필요하며 데이터가 분산되는 방식 때문에 배열 드라이브의 전체 용량이 하나로 통합된다. 

- 예를들면 스트라이프 RAID 0 구성에서 쌍을 이룬 2개의 1TB드라이브는 하나의 2TB 볼륨으로 인식된다. 하지만 RAID 0는 패리티(parity)데이터를 미러링 처리하거나 저장하지 않아서 디스크에 문제가 생기면 나머지 디스크에 저장된 데이터가 무용지물이 된다는 단점이 있다.

#### 2-2. RAID 1 (mirroring)

- RAID1 또는 미러링 또한 최소한 2개의 드라이브가 있어야한다. 하지만 데이터를 스트라이핑하는 대신, 한 드라이브에 저장된 데이터가 다른 배열 드라이브로 복제된다. RAID1은 일정 수준의 데이터 중복을 제공하며, 드라이브 고장시 데이터를 잃지 않고도 배열을 다시 구성할 수 있다.

- 최소한 HDD를 사용할 경우 성능 향상을 경험할 수 있고, 읽고자 하는 데이터에 더 가까운 배열 드라이브 헤더가 더 빨리 데이터에 접근하므로 탐색시간과 회전 대기 시간이 줄어든다. 단점으로 데이터가 배열의 모든 드라이브에 미러링 처리가 되어서 일반적으로 성능이 저하된다. 
- RAID1의 전체 용량은 중복으로 인해 단일 드라이브의 용량과 동일하다. 2개의 1TB드라이브를 사용하면 RAID1의 전체 용량은 여전히 1TB이다.

#### 2-3. RAID 3 및 RAID4

- 패리티가 단순제공되는 스트리핑된 세트(적어도 3개의 디스크)
- RAID4는 2개의 디스크

#### 2-4. RAID 5 

- 패리티가 배분되는 스트리핑된 세트(적어도 3개의 디스크)

#### 2-5. RAID 6

- 패리티가 배분되는 스트리핑된 세트(적어도 4개의 디스크)


#### RAID 0+1
- 레이드 0+1은 먼저 디스크를 스트리핑(RAID 0)하고, 디스크를 미러링(RAID 1) 한다. (적어도 4개의 디스크) 디스크가 6개일경우는 3개씩 스트리핑하고 미러링을 그다음에 수행한다.

#### RAID 10(RAID 1+0)
- 레이드 10은 먼저 디스크를 미러링(RAID 1)하고, 그 이후 스트리핑 한다. (적어도 4개의 디스크)
디스크가 6개일 경우는 2개씩 미러링을 하고, 미러링된 3개를 스트리핑 한다.

### 3. 페일오버(Failover)와 페일백(Failback)의 개념정리 및 차이점 분석

#### 3-1.failover
- 시스템에 이상이 생겼을때 예비 시스템으로 자동전환 되는 기능
- 전원장치, 네트워크 장비, 서버등 이중화 시스템에 적용됨
- active-standby 구조에서 Active 장애시 standby로 자동전환되는 기능

#### 3-2. failback
- failover에 따라 전환된 운용환경을 장애 전 상태로 되돌리는 처리를 말한다.

페일오버: 0 1 2 error 3 4 5
페일백: 0 1 2 error 2 3 4
