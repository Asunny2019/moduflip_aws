## 인프라운영 따라잡기 3주차 자율학습  

### 1. Xeon프로세서의 종류 및 특성
![](https://github.com/Asunny2019/moduflip_aws/blob/main/moduflip17_aws/img/3rd_1.jpg?raw=true)  


제온 프로세서는 워크스테이션, 서버등 특별한 용도의 컴퓨터를 위해 개발된 CPU로 일반 PC용으로 개발된 코어 i 프로세서 등과 다른 특성을 갖춘 CPU이다. 제온 프로세서는 가상화 (VM), 병렬처리, 멀티코어, 신호 암호화, ECC(오류검출, 서버는 메모리 사용이 잦기 때문에 오류 발생확률이 높다. 때문에 오류를 찾고 수정하는 기능이 필요) 메모리 지원, 다중 CPU 소켓 등 워크스테이션이나 서버를 위한 기능을 제공한다.


과거 제온 프로세서는 크게 E3, E5, E7, D등 네가지 모델로 나눌 수 있다.
모델을 나누는 기준은 성능과 용도이다. 이름 뒤에 붙는 숫자 4자리는 구체적인 모델명이고 V(숫자)는 버전을 의미한다. 즉 '제온 E5-2600 V4'는 제온 E5 프로세서의 4번째 모델이다.


a. E3
제온 프로세서 중에서 가장 하위 모델로 싱글 소켓서버와 개인용 워크스테이션을 위한 제품이다. 사실 E3의 가장 큰 특징은 순수한 제온 프로세서가 아니라는 점이다. 일반 PC용 프로세서인 코어 i 시리즈에서 내장 그래픽 프로세서를 제거하고 서버 워크스테이션을 위한 기능 일부를 추가했다. 그래서 과거에는 코어 i 시리즈용 메인보드에 제온 E3을 꽂아서 사용할 수도 있었지만 지금은 불가능하다.

 b. E5
 제온 프로세서의 중급 모델로 가장 많이 판매되는 제품으로 중대형 서버와 클라우드 데이터 센터를 위한 프로세서이다. 제온 프로세서 전체 판매량의 70% 를 제온 E5가 차지하고 있다. 실제로 대부분의 서버는 제온 E5를 활용해 제작되고 클라우드 데이터 센터에서 가장 많이 공급되는 모델이기도 하다. 싱글 또는 듀얼소켓을 지원한다. 일반 사용자는 E3, E5 모델만 구매 가능하다.
 
 c. E7
 제온 프로세서의 최고급 모델이다. 시중에 존재하는 단일 프로세서 가운데 최고 성능을 자랑한다. 미션크리티컬, 슈퍼컴퓨터, 대용량 컴퓨터을 위한 스케일 업등에 활용된다. 하나의 보드에 최대 8개의 프로세서를 연결해 병렬 컴퓨팅을 구성할 수 있다. 일반 사용자는 구매할 수 없고 기업이 대량으로 구매를 원할 때 인텔이 특별 공급을 한다.
 
d. D
 D는 저전력 서버를 구축하기 원하는 기업을 위한 모델로 아톰프로세서를 기반으로 제작되어 전력 소모량이 적고 병렬 컴퓨팅을 통해 만족할 만한 성능을 낸다.

오늘날은 용도에 따라 모델명을 나눈다.

#### 인텔 제온 E 프로세서
비지니스에 적합한 필수성능과 엔트리급 서버 솔루션을 위한 확장성과 안정성을 갖춘 모델

#### 인텔 제온 W 프로세서
크리에이티브 전문가를 위해 설계되었으며 워크스테이션에서 VFX, 3D 렌더링 및 3D CAD에 필요한 성능을 제공한다.

#### 인텔 제온 D 프로세서
데이터 센터에서 인텔리전트 에지에 이르는 공간 및 전력 제약이 있는 환경에서 워크로드에 최적화된 성능을 제공한다.

### 3. XeonGold5115프로세서의 캐시구조 살펴보기

intel xeon gold 5115프로세서의 캐시구조는 여러 단계로 구성되어있다. 이 아키텍처에서는 L1, L2 캐시에 있는 데이터가 L에도 존재한다. 이 때문에 다른 코어는 자신 이외의 캐시를 확인하지 않고 L3캐시만 확인하면 되는 구조이다. 
왜 메모리 영역이 몇개나 존재하는 것일까? 메모리를 이용하려면 메모리 컨트롤러를 경유해서 일단 CPU 밖으로 나가야 한다. 고속 CPU에서는 이런 처리 지연(Latency)를 허락하지 않는다. 처리 지연을 줄이기 위해서 가장 자주 사용하는 명령/데이터를 코어 가까운 곳에 배치하는 것이다. 

영역이 여러 단계로 나누어져 있는 이유는 엑세스 속도 때문이다. 일반적으로 캐시 메모리가 커질수록 액세스 속도가 느려진다. 하지만 가능한 CPU 가까운 곳에 많은 캐시를 두고 싶은 것도 사실이다. 이 때문에 캐시를 여러 단계로 배치해서 초고속으로 엑세스하고 싶은 데이터는 L1 캐시에 준고속으로 엑세스하고 싶은 데이터는 L2캐시에 두는 형태로 만드는 것이다.
