## 6주차 자율학습

### 1. Polkit(Policykit)의 특징 및 적용방법, 사용이 필요한 경우 생각해보기
    - 유닉스계열 OS에서 시스템 전반에서 인증 단계를 통제하기 위해 만들어진 프로그램으로 인증을 받지 않은 프로세스로 하여금 인증을 받은 계정과의 커뮤니케이션이 가능하도록 체계적인 방법을 제공함.
    - 특징: sudo 등 기존의 인증 시스템과 달리 모든 프로세스에 대하여 최고 관리자의 권한을 요구하는 것이 아니라 집중화된 시스템 정책의 통제아래에서 보다 느슨한 수준의 인증 수행한다.
    - 사용이 필요한 경우: 시스템에 큰 위해를 가하지 않는 수준에서 자주 사용해야하는 프로세스를 자주 진행하는 경우에 Polkit을 많이 사용하게 될 것 같다. 예를들어 usb를 마운트하는 과정등....

### 2. 데몬 프로세스의 특징 및 자주 사용되는 데몬 프로세스 5가지 생각하기
    - 데몬(demon)이란 서비스의 요청에 대해 응답하기 위해 오랫동안 실행중인 백그라운드(background)프로세스이다. 유닉스(리눅스 포함) 운영체제에서 이름이 'd'로 끝나는 프로세스들이 대표적으로 inetd, httpd, nfsd, sshd, named,lpd등이 있다. 
    - 여기서 핵심은 "백그라운드 프로세스"라는데 있다. 
    - 백그라운드 프로세스의 반대개념으로 포그라운드(foreground) 프로세스가 있다는 것도 유추할 수 있는데 포그라운드 프로세스는 사용자와 대화창구인 표준 입출력장치(터미널과 키보드)를 통해 대화한다. 
    - __백그라운드 프로세스는 적어도 입력장치에 대해 터미널과의 관계를 끊은 모든 프로세스__ 를 의미한다. 사용자에게 무언가를 키보드를 통해 전달 받지 않고 __스스로 동작하는 프로세스__ 가 바로 백그라운드 프로세스이다. 
    - 데몬 프로세스란?: 백그라운드 프로세스 중에서 부모프로세스가 1 혹은 다른 데몬 프로세스인 프로세스를 말한다. 
    
    - 가장 대표적인 데몬 프로세스: 웹서버, 데이터 베이스 서버, 원격접속 서버 등....

### 3. 리눅스에서 사용되는 주요 파일 시스템의 종류에 관해서 정리하기(Tip: xfs, ext4 등)  

#### a. XFS : 대용량 파일시스템(64비트 파일시스템으로 900만 TB까지 지원한다.)  
__장점__
    - 저널링:파일수에 관계 없이 예상치 못한 상황으로부터 신속한 복구 및 재시작
    - 신속한 트랜잭션: 읽기/ 쓰기
    - 높은 확장성
    - 파일 시스템 체크속도가 사용중인 inode 만을 대상으로 하기 때문에 빠름
    - 동일조건의 파일 삭제시간 또한 EXT3에 비해 2.5배 가량 빠르다. 
    - 안정적이다. 많은 상용사이트에서 검증 됨
        
__단점__
    - 작은 파일에서 느림(작은 파일은 JFS파일 시스템에서 더 좋은 성능을 발휘한다)
    - 파일 시스템 확장은 가능하나 축소는 안된다. 어짜피 데이터 무결성을 위해 백업을 해야하므로 다른 파일시스템에서도 축소는 무의미함.
    - 이식성이 떨어진다.
    - 파일시스템에 문제가 생겼을 경우 커널메시지가 제대로 안올 수 있음.
    - 디버그 옵션 없음. 삭제된 파일은 복구 불가능
    - 별도의 데몬 관리를 해야함
        
#### b. ext4: 64비트 기억공간 제한을 없애고 ext3의 성능을 향상시키며, 하위 호환성이 있는 확장버전으로 클러스터 파일 시스템 사에서 개발 되었다. 다른 커널 개발자들이 ext3사용자에게는 영향을 주지 않으면서 ext3에서 fork하여 ext4로 이름변경을 제안하여서 ext4 개발 계획 발표했다.

__장점__
    - 대형파일 시스템: 최대 1엑사 바이트의 볼륨과 최대 16테라 바이트의 파일을 지원한다.
    - 하위호환성: ext3와 ext2에 대한 하위 호환성이 있어서 ext3과 ext2 파일 시스템을 ext4로 마운트하는 것이 가능하다.
    - 지연된 할당: allocate-on-flush라는 파일 시스템 성능기술사용
    - 32000개의 하위 디렉토리 제한 없음.
    
__단점__
    - 지연된 할당과 데이터 유실 가능성: 지연된 할당은 프로그래머가 ext3에서 의도했던 동작을 변경하기 때문에 모든 데이터가 디스크에 기록되기 전에 발생한 시스템 충돌이나 전원 차단시 추가적인 데이터 유실 위험을 야기한다. 2.6.30이상의 커널에서는 자동으로 이런 경우를 탐지해서 이전의 동작으로 되돌린다. 

#### c. Btrfs

 - 파일 최대 크기: 최대 16엑사 바이트의 볼륨과 최대 16엑사 바이트의 파일을 지원한다. (ext4보다 큰 파일최대 크기 지원)
 - innode 할당방식 EXT4는 파일시스템을 생성하면서 파일 시스템이 지원하는 최대 inode개수를 정의한다. 파일 시스템이 생성된 후에는 지원 가능한 최대 개수를 변경할 수 없어서 작을 크기의 파일을 많이 생성하면 파일 시스템의 free space가 많이 남더라도 free inode가 부족해서 파일 생성이 불가능함. 반면에 Brt는 유연한 inode할당방식을 갖고 있어서 필요한 만큼 inode의 개수를 추가할 수 있다. 
 - EXT4는 저널링 파일시스템이지만 COPY-on-Write는 지원하지 않는다. 
 - Btrfs는 Copy-on-Write시스템이지만 저널링을 지원하지 않음. 블록 변경사항을 새로운 위치에 기록한 다음 link를 변경한다. write가 끝나기 전까지는 변경사항을 commit하지 않음. 기존데이터가 수정되지 않아서 쓰기 중에 에러가 발생되어도 기존 데이터는 손상 안됨.
