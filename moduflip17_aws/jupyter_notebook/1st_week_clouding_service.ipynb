### 클라우드 컴퓨팅(cloud computing)

그리드 컴퓨팅, 유틸리티 컴퓨팅, 서비스로서의 소프트웨어 개념이 혼합된 컴퓨팅 인프라의 집합이다. 데스크탑PC, 휴대폰, 노트북등 물리적으로 서로 다른 위치에 존재하는 다양한 정보들을 웹기반 애플리캐이션을 활용하여서 대용량 데이터 베이스를 가상화 기술을 통합하여 제공하는 기술 및 환경을 말한다. 즉, 인터넷 기반 컴퓨팅의 일종으로 정보를 자신의 컴퓨터가 아닌 인터넷에 연결된 다른 컴퓨터로 처리하는 기술을 말한다. 컴퓨터의 각종 연산 및 처리를 개인 PC가 아닌 인터넷 공간에서 사용하는 것이다.


__클라우드 컴퓨팅의 장점__

1. __유연한 IT인프라 관리__ : 실시간으로 IT인프라를 조정가능
2. __신속한 인프라 도입__ : 클라우드 서비스 가입후 몇분 이내에 서비스 구축가능
3. __간편한 글로벌 서비스__ : 과거에는 글로벌 서비스를 위해서 기업이 특정 대륙과 지역에 데이터 센터를 구축해야했지만 클라우드 사엽자가 미리 구축한 데이터 센터 활용
4. __예상치 못한 트래픽 폭주 대응__: 트래픽 폭주에 맞춰 기업이 별도의 인프라를 구축하지 않아도 알아서 인프라가 늘었다 줄어드는 자동 트래픽 증감기술인 __'오토 스케일링'__ 이 개발 되어 있어서 빠르게 대응할 수 있다. 
5. __빅데이터 인공지능 서비스 확장__: 개인의 스마트 기기에서 인공지능을 구현하는 것은 불가능하지만 중앙 클라우드에서 인공지능을 구축하고 이를 사용자가 이용하게 하면, 사용자는 중앙 서버의 컴퓨팅으로 간편하게 인공지능 서비스를 받을 수 있다.

클라우드 컴퓨팅 서비스를 설명하기에 앞서 전통적인 IT서비스 방식인 __온프레미스(On-premise)__ 를 언급하고자 한다.
온프레미스는 기업이 자체적으로 데이터 센터를 보유하고 시스템구축부터 운영까지 모두 수행하는 형태를 뜻한다.

온프레미스 기반으로 IT 서비스를 준비한다면, 최소한 다음 같은 사항이 필요하다.

- 전산실 구성, 운영체제 선택과 구매, 데이터베이스 선택과 구매, 최대 사용량에 맞추어 서버 구매, 인터넷 환경을 위한 전용선 설치 및 구매, LAN 또는 스위칭 등 네트워크 구축, 방화벽 등 보안 시스템 구축, 운영체제와 데이터베이스 등 서버구성, 서비스용 소프트웨어 구매 및 개발, 추가 업그레이드 모듈 설치, 단계별 백업 구성 및 설정, 모니터링 시스템 구성, 장애 대비를 위한 시스템 구성

이렇게 구축한 이후에도 운영, 유지 보수가 필요하기 때문에 위의 클라우드 컴퓨팅에 비해 구축 및 유지비용이 많이 든다.



### 1. IaaS,  PaaS,  SaaS의  비교

IaaS(Infrastructure-as-a-Service), Paas(Platform as a Service), SaaS(Software as a Service)는 클라우드 컴퓨팅 서비스의 유형이다.

초기 클라우드 서비스는 소프트웨어를 웹에서 쓸 수 있는 __SaaS__ (구글시트, 드롭박스 등)가 대부분이었으나, 이후 서버와 스토리지, 네트워크 장비 등의 IT인프라 장비를 빌려주는 __IaaS__ 와 플랫폼을 빌려주는 __PaaS__ 으로 확장되고 있다.

클라우드 서비스는 어떤 자원을 제공하느냐에 따라서 __SaaS__, __IaaS__, __PaaS__ 세가지로 나뉘게 된다. 
- IaaS(Infrastructure-as-a-Service)
: 서드파티 업체가 제공하는 고도로 자동화되고 확장 가능한 IT인프라를 의미한다. 이 인프라에는 스토리지, 호스팅, 컴퓨팅, 네트워킹이 포함된다.
